Part 1: Simple Charts

1. Pie Chart: 
Make a pie chart that shows the distribution of animals in a zoo that contains
30 penguins, 40 polar bears, 20 lions, and 10 otters

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
labels = ['penguins','polar bears','lions','otters']
count = [30,40,20,10]
colors = ['grey', 'brown', 'yellow','pink']
explode = [0,0.1,0,0] #explode 2nd slice
plt.pie(count, explode=explode, labels=labels, colors=colors,autopct='%1.0f',shadow=True,startangle=0)
plt.title('Distribution of animals in a zoo',fontweight='bold')
plt.axis('equal')
plt.show()
```

2. Bar Chart:
Make a bar chart that shows the number of people, out of a total of 40, who like various movie types
if 10 people like comedy; 8 people like action;
15 people like romance; 2 people like drama, and 5 people like sci-fi

```python
genre = ('Comedy', 'Action', 'Romance', 'Drama', 'Sci-fi')
y_position = np.arange(len(genre))
count = [10,8,15,2,5]
rect=plt.bar(y_position, count, align='center', alpha=0.6, color='lightblue')
plt.xticks(y_position, genre, rotation='0')
plt.ylabel('Number of People')
plt.xlabel('Genres')
plt.title('People and likes for movies')

#for i in range(len(count)): 
 #   plt.annotate(str(count[i]), xy=(genre[i],count[i]))

#def autolabel(rects): """ Attach a text label above each bar displaying its height """ 
for rects in rect: 
    height = rects.get_height()
    plt.text(rects.get_x() + rects.get_width()/2., 1*height, str(height), ha='center', va='bottom')
#autolabel(rects1) 
#autolabel(rects2)

plt.show()
```

Part 3: 3D Plots

In your experiments you found that the phenomenon you are measuring is described by the following equation:
Z = np.exp(-((X-1)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-3)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-2)**2+(Y-3)**2)/0.15) + \
    np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)
Note: For your convenience, the python version of the equation is written with np.exp referring to the exponentiation function ex from the numpy package name shortened to np (like in the tutorial). Also, ** is the power function. The \ is because python needs it for equations that continue over multiple lines - not necessary in other languages.
Also, to be complete, be sure to show the contour plot for the entire range for which this function has interesting features to observe.

1. Contour plot

1). Make a contour plot. Make sure to add labels in the plot or a legend for colors on the contours. 
You can choose either a filled contour plot or colored lines, your choice.

```python
fig = plt.figure(figsize=(5,4))
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes([left, bottom, width, height])
start, stop, n_values = -10,10,100
x_vals = np.linspace(start, stop, n_values)
y_vals = np.linspace(start, stop, n_values)
X, Y = np.meshgrid(x_vals, y_vals)
Z = np.exp(-((X-1)**2+(Y-4)**2)/0.15) + np.exp(-((X-3)**2+(Y-4)**2)/0.15) + np.exp(-((X-2)**2+(Y-3)**2)/0.15) + np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)
cont_plot = plt.contourf(X, Y, Z)
plt.colorbar(cont_plot)
ax.set_title('Contour Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

2). Do this for an additional color mapping: e.g. hot/cold or black/white.

```python
cont_plot1 = plt.contourf(X, Y, Z,cmap=mplot.cm.afmhot)
plt.colorbar(cont_plot1)
ax.set_title('Contour Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
