Steps:
1. Input the following data into a data frame called titanic, and display the entire data frame:
 Sex, Class, Survived, Died
 Children, First, 6, 0
 Children, Second, 24, 0
 Children, Third, 27, 52

```python
# defininig input dictionary
data= {
    "Sex" : ["Children","Children","Children","Men","Men","Men","Men","Women","Women","Women","Women"],
    "Class" : ["First","Second","Third","First","Second","Third","Crew","First","Second","Third","Crew"],
    "Survived" : [6,24,27,57,14,75,192,140,80,76,20],
    "Died" : [0,0,52,118,154,387,693,4,13,89,3]
}
Titanic = pd.DataFrame(data)
Titanic
```

2. Now only show the data of the people in first class.

```python
Titanic.loc[Titanic["Class"]=="First"]
```

3. Delete the crew members from the data.

```python
dropped_df=Titanic.drop([6,10])
dropped_df
```

4. Create a new column that is the total number of people for that group (those who survived + died).

```python
dropped_df["total"]=dropped_df["Survived"]+dropped_df["Died"]
dropped_df
```

5. Create a new column with the percentage of people who survived.

```python
dropped_df["percentage"]=dropped_df["Survived"]/dropped_df["total"]*100
dropped_df
```

6. Delete the column indicating the total number of people in that group.

```python
dropped_df.drop("total",axis=1,inplace=True)
dropped_df
```

7. Only show the rows where more than 80% of the people survived.

```python
dropped_df.loc[dropped_df["percentage"]>80]
```

8. Then only show the rows where less than 40% of the people survived.

```python
dropped_df.loc[dropped_df["percentage"]<40]
```

9. Calculate the total number of people that survived and died for each class, then report the percentages.
(Hint: Use a grouped calculation.)

```python
classified=dropped_df.groupby('Class').sum()
classified.drop("percentage",axis=1,inplace=True)
classified["total"]=classified["Survived"]+classified["Died"]
classified["Percentage_of_Survivors"]=classified["Survived"]/classified["total"]*100
classified["Percentage_of_Dead"]=classified["Died"]/classified["total"]*100
classified
```
